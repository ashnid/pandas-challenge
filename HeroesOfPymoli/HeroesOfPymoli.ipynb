```python
# Dependencies and Setup
import pandas as pd

# File to Load
file_to_load = "purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count
- Display the total number of players

```python
# Value_count counts the number of unique values in a column
count_unique = purchase_data["SN"].nunique()
#count_unique
```

```python
#Display the total number of players
total_players_df = pd.DataFrame([
    {"Total Players": count_unique}
])
total_players_df
```

## Purchasing Analysis (Total)
- Number of Unique Items
- Average Purchase Price
- Total Number of Purchases
- Total Revenue

```python
#Number of unique items
count_unqiue_items = len(purchase_data["Item ID"].value_counts())
#count_unqiue_items 
```

```python
#Average purchase price 
average_price = purchase_data["Price"].mean()
#average_price
```

```python
#Total Number of Purchases
total_number_purchases =  purchase_data["Purchase ID"].nunique()
#total_number_purchases
```

```python
#Total Revenue
total_rev = purchase_data["Price"].sum()
#total_rev
```

```python
# Create a dataframe for the Purchasing Analysis data
purchasing_analysis_df = pd.DataFrame ([
    {"Number of Unique Items": count_unqiue_items,
     "Average Price":average_price, 
     "Number of Purchases": total_number_purchases,
     "Total Revenue": total_rev}
])
purchasing_analysis_df.style.format({"Average Price":"${:,.2f}",
                                "Total Revenue":"${:,.2f}"})
```

## Gender Demographics
- Percentage and Count of Male Players
- Percentage and Count of Female Players
- Percentage and Count of Other / Non-Disclosed

```python
#Group purchase date by gender
groupby_gender = purchase_data.groupby("Gender")

#Count the unique values in the data
#Number of unique items in each group under SN
gender_count = groupby_gender.nunique()["SN"]

#Find percentage by dividing by total number of players
percentage_of_players = gender_count / count_unique * 100
```

```python
#Create dataframe for gender demographics data
gender_demo = pd.DataFrame({"Percentage of Players": percentage_of_players, 
                            "Total Count": gender_count })
gender_demo.style.format({"Percentage of Players":"{:.2f}%"})
```

## Purchasing Analysis (Gender)
The below each broken by gender
   - Purchase Count
   - Average Purchase Price
   - Total Purchase Value
   - Average Purchase Total per Person by Gender

```python
#Find Purchase Count by counting the number of values in the Purchase ID column of the grouped data
purchase_count = groupby_gender["Purchase ID"].count()

# Average Purchase Price
avg_price = groupby_gender["Price"].mean()

# Total Purchase Value
avg_total = groupby_gender["Price"].sum()

# Average Purchase Total per Person by Gender 
avg_per_person = avg_total/gender_count
```

```python
purchasing_analysis = pd.DataFrame({"Purchase Count": purchase_count, 
                                    "Average Purchase Price": avg_price,
                                    "Total Purchase Value": avg_total,
                                    "Avg Purchase Total per Person": avg_per_person})
purchasing_analysis.style.format({"Total Purchase Value":"${:,.2f}",
                                  "Average Purchase Price":"${:,.2f}",
                                  "Avg Purchase Total per Person":"${:,.2f}"})
```

## Age Demographics
The below each broken into bins of 4 years (i.e. <10, 10-14, 15-19, etc.)
    - Purchase Count
    - Average Purchase Price
    - Total Purchase Value
    - Average Purchase Total per Person by Age Group

```python
# Create bins
# Bins are 0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 99999
bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 99999]

#Create names for bins
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
```

```python
# Put data into bins
purchase_data ["Age Group"] = pd.cut(purchase_data["Age"], bins,labels=group_names, include_lowest = True)

#Create group based on the bins
grouped_by_age = purchase_data.groupby("Age Group")

# Count total players by age 
# Number of unique names in SN column
total_count_age = grouped_by_age["SN"].nunique()

# Calculate percentages by age category 
percentage_by_age = (total_count_age/count_unique) * 100

#Create dataframe for Age Demographics
age_demo = pd.DataFrame({"Percentage of Players": percentage_by_age, 
                            "Total Count": total_count_age })
age_demo.style.format({"Percentage of Players":"{:.2f}%"})
```
